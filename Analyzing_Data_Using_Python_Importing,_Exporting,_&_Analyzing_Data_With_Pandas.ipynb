### Computing Unique Values and Renaming Columns

```python
superstore_data['Region'].unique()
```

```python
superstore_data['Region'].value_counts()
```

```python
superstore_data.columns
```

```python
superstore_data.rename(columns={'Sales':'Total Sales'}, inplace=True)

superstore_data.head()
```

```python
superstore_data.rename(columns={
    'Product Name':'Product',
    'Quantity':'Product Quantity'}, inplace=True)

superstore_data.head()
```

```python
superstore_data.rename(columns=str.upper,inplace=True)
superstore_data
```

```python
superstore_data.columns = [column.replace(' ','_') for column in superstore_data.columns]
```

### Sorting Data Based on Specific Columns

```python
superstore_data.sort_values('CUSTOMER_NAME', ascending=True, inplace=True)
superstore_data.head()
```

```python
superstore_data.sort_values('TOTAL_SALES', inplace=True, kind='mergesort')
superstore_data.head(10)
```

```python
superstore_data.sort_values('CUSTOMER_NAME', ascending=False, inplace=True)
superstore_data.head()
```

```python
superstore_data.sort_values(['CUSTOMER_NAME','TOTAL_SALES'],
                            ascending=[True,False],inplace=True)
superstore_data.head(5)
```
