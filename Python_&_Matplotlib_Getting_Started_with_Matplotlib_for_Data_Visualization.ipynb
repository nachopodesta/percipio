### Visualizing Time Series Data Using Line Charts

```python
import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
stock_price_ba = pd.read_csv('/content/drive/MyDrive/COLAB NOTEBOOKS/Percipio/BA.csv')
stock_price_ba.head()
```

```python
stock_price_ba.describe()
```

```python
stock_price_ba.dtypes
```

```python
stock_price_ba['Date'] = pd.to_datetime(stock_price_ba['Date'], format='%Y-%m-%d')
```

```python
stock_price_ba.dtypes
```

```python
plt.figure(figsize=(12,8))

plt.plot(stock_price_ba['Date'], stock_price_ba['Close'], label="Closing Stock Price")

plt.xlabel('Date', fontsize=12, color='Blue')
plt.ylabel('Price', fontsize=12, color='Blue')

plt.legend()

plt.title('Boing Stock Price (2015-2020)')
```

### Customizing Ling Chart Elements

```python
plt.figure(figsize=(12,8))

plt.plot(stock_price_ba['Date'], stock_price_ba['Open'], label="Opening Stock Price")
plt.plot(stock_price_ba['Date'], stock_price_ba['Close'], label="Closing Stock Price")

plt.xlabel('Date', fontsize=12, color='Blue')
plt.ylabel('Price', fontsize=12, color='Blue')

plt.legend()

plt.title('Boing Stock Price (2015-2020)')
```

```python
plt.figure(figsize=(12,6))

plt.plot(stock_price_ba['Date'], stock_price_ba['Low'], color='Red',
         label='Daily lows')

plt.plot(stock_price_ba['Date'], stock_price_ba['High'], color='Green',
         label='Daily highs')

plt.xlabel('Date', fontsize=15, color='Blue')
plt.ylabel('Price', fontsize=15, color='Blue')

plt.legend()

plt.title('Boeing Stock Price (2015-2022)', fontsize=20)
```

```python
plt.figure(figsize=(12,8))

plt.plot(stock_price_ba['Date'], stock_price_ba['Close'],
         color='Purple', label='Closing Stock Price', linestyle='--', linewidth=2)

plt.xlabel('Date', fontsize=15, color='Blue')
plt.ylabel('Price', fontsize=15, color='Blue')

plt.legend()

plt.title('Boeing Stock Price (2015-2022)', fontsize=20)
```

```python
plt.figure(figsize=(12,8))

plt.plot(stock_price_ba['Date'], stock_price_ba['Volume'])

plt.tick_params(axis='both', color='Blue',labelcolor='Blue',labelsize='large')

plt.xlabel('Date', fontsize=15, color='Blue')
plt.ylabel('Volume TradedPrice', fontsize=15, color='Blue')

plt.legend()

plt.title('Boeing Volume Traded (2015-2022)', fontsize=20)
```

### Creating Line Charts Using the Twinx() Function

```python
plt.figure(figsize=(12,8))

plt.plot(stock_price_ba['Date'], stock_price_ba['Close'], label='Boing')
plt.plot(stock_price_msft['Date'], stock_price_msft['Close'], label='Microsoft')

plt.xlabel('Date', fontsize=12, color='Blue')
plt.ylabel('Price', fontsize=12, color='Blue')

plt.legend()
plt.title('Boing and Microsoft Stock Price (2015-2020)')
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(12,8))

ax1.plot(stock_price_ba['Date'], stock_price_ba['Close'], label='Boeing', color='red')
ax1.grid(alpha=.4)

ax2 = ax1.twinx()

ax2.plot(stock_price_msft['Date'], stock_price_msft['Close'], label='Microsoft', color='green')
ax2.set_title('Microsoft vs. Boeing 2015-2020', fontsize=22)

plt.show()
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(12,8))

ax1.plot(stock_price_msft['Date'], stock_price_msft['Close'], label='Close', color='orange')
ax1.grid(alpha=.4)

ax1.legend()

ax2= ax1.twinx()
ax2.plot(stock_price_msft['Date'], stock_price_msft['Volume'], label='Volume', color='purple')
ax2.set_title('Microsoft Close and Volume 2015-2020', fontsize=22)

ax2.legend()
plt.show()
```

### Creating Lollipop Charts

```python
loan_data.head()
```

```python
loan_data['Age_bucket'] = pd.cut(loan_data['Age'],8)
loan_data.head()
```

```python
load_data_age_count = loan_data.groupby('Age_bucket').count()
load_data_age_count
```

```python
categories = loan_data['Age_bucket'].cat.categories
categories
```

```python
index = np.array([x for x, _ in enumerate(categories)])
index
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(index, load_data_age_count['ID'], color='firebrick', width=0.4)

plt.xticks(index, categories)
plt.xticks(rotation=90)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.vlines(x=index, ymin=0, ymax=load_data_age_count['ID'], color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=index, y=load_data_age_count['ID'], s=75, color='firebrick', alpha=0.7)
```

### Creating Charts That Visualize Multiple Variables

```python
loan_data_age_mean = loan_data.groupby('Age_bucket').mean()
loan_data_age_mean
```

```python
fig, ax = plt.subplots(figsize=(12,6))

width=0.4

ax.bar(index, loan_data_age_mean['Loan_length'], color='firebrick', width=0.4, label="Lenght")
ax.bar(index+width, loan_data_age_mean['Signers'], color='yellow', width=0.4, label="Signers")

plt.xticks(index+width/2, categories)
plt.xticks(rotation=90)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

offset=0.4

ax.vlines(x=index, ymin=0, ymax=loan_data_age_mean['Loan_length'],
          color="firebrick", alpha=0.7, linewidth=2, label='Length')

ax.scatter(x=index, y=loan_data_age_mean['Loan_length'], s=75, color="firebrick", alpha=0.7)

ax.vlines(x=index+offset, ymin=0, ymax=loan_data_age_mean['Signers'],
          color="yellow", alpha=0.7, linewidth=2, label='Signers')

ax.scatter(x=index+offset, y=loan_data_age_mean['Signers'], s=75, color="yellow")

plt.xticks(index+offset/2, categories)
plt.xticks(rotation=90)
plt.legend()
```
